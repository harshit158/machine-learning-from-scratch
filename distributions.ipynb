## PMF (Probability Mass Function)

Discrete variables

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the outcomes and their probabilities
outcomes = np.array([1, 2, 3, 4, 5, 6])
probabilities = np.array([1/6] * 6)  # Each outcome has a probability of 1/6

# Create the PMF plot
plt.figure(figsize=(10, 5))
plt.bar(outcomes, probabilities, color='blue', alpha=0.7)
plt.xticks(outcomes)
plt.ylim(0, 0.25)
plt.title('Probability Mass Function (PMF) of a Fair Six-Sided Die')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.grid(axis='y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the outcomes and their probabilities for a six-sided die
outcomes = np.array([1, 2, 3, 4, 5, 6])
probabilities = np.array([1/6] * 6)  # Each outcome has a probability of 1/6

# Calculate the CDF by taking the cumulative sum of probabilities
cdf = np.cumsum(probabilities)

# Create the CDF plot
plt.figure(figsize=(10, 5))
plt.step(outcomes, cdf, where='post', color='blue', lw=2)
plt.xticks(outcomes)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.title('Cumulative Distribution Function (CDF) of a Fair Six-Sided Die')
plt.xlabel('Outcome')
plt.ylabel('Cumulative Probability')
plt.grid()
plt.ylim(0, 1)
plt.show()
```

## PDF (Probability Density Function)

Continuous variables

```python
from scipy.stats import norm

# Generate x values from -4 to 4
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, loc=0, scale=1)  # Standard normal distribution with mean=0, std_dev=1

# Create the PDF plot
plt.figure(figsize=(10, 5))
plt.plot(x, pdf, color='green', lw=2)
plt.fill_between(x, pdf, alpha=0.2, color='green')
plt.title('Probability Density Function (PDF) of a Standard Normal Distribution')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.xlim(-4, 4)
plt.ylim(0, 0.45)
plt.show()
```

```python
from scipy.stats import norm

# Generate x values from -4 to 4
x = np.linspace(-4, 4, 1000)
cdf = norm.cdf(x, loc=0, scale=1)  # Cumulative distribution function for standard normal

# Create the CDF plot
plt.figure(figsize=(10, 5))
plt.plot(x, cdf, color='green', lw=2)
plt.fill_between(x, cdf, alpha=0.2, color='green')
plt.title('Cumulative Distribution Function (CDF) of a Standard Normal Distribution')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid()
plt.xlim(-4, 4)
plt.ylim(0, 1)
plt.show()
```
